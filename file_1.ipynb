2. PROJECT SUMMARY - 

    The Tourism Experience Analytics project is an end-to-end machine learning application that analyzes tourist behavior and delivers predictive insights and attraction recommendations through an interactive Streamlit web app.


    The system includes two supervised learning tasks:

    - Rating Prediction (Regression): A CatBoost Regressor predicts attraction ratings using contextual features and engineered aggregate statistics such as attraction-level averages and seasonal trends.
    - Visit Mode Prediction (Classification): A CatBoost Classifier predicts the likely visit mode (e.g., Family, Friends, Solo) based on geographic, attraction, temporal, and rating-related features.


    Additionally, an item-based collaborative filtering recommendation system suggests attractions using similarity patterns derived from user–attraction interactions (“People who visited X also visited Y”).
    
    The project demonstrates a complete ML pipeline consisting of data preprocessing, feature engineering, model training, artifact management and deployment, integrated into a single interactive application.

3. PROBLEM STATEMENT -
    
    Tourism agencies and travel platforms aim to enhance user experiences by leveraging data to provide personalized recommendations, predict user satisfaction, and classify potential user behavior. This project involves analyzing user preferences, travel patterns, and attraction features to achieve three primary objectives: regression, classification, and recommendation.

We will be creating two dataframes from the attr dataframe. One will contain only int64 objects in the "AttractionTypeId" column and the other column will have only string objects in the "AttractionTypeId" column. This will help clean the whole attr dataframe properly. After cleaning, we will merge both of them two form the original attr dataframe again.

```python
# Creating attr_num that will contain int64 objects in the AttractionTypeID column
attr_num = attr[pd.to_numeric(attr['AttractionTypeId'], errors='coerce').notna()]

# Enforcing the dtype of the AttractionTypeID column to be int64
attr_num['AttractionTypeId'] = attr_num['AttractionTypeId'].astype(int)

# Checking few rows of the new dataframe
attr_num.sample(7)
```

```python
# Creating attr_str that will contain string objects in the AttractionTypeID column
attr_str = attr[pd.to_numeric(attr['AttractionTypeId'], errors='coerce').isna()]

# Checking few rows of the new dataframe
attr_str.sample(7)
```

```python
# Checking the shape and info of the attr_str dataframe
print(attr_str.shape)
print(attr_str.info())
```

The sample of attr_str from above shows that the "AttractionTypeID" column has categorical entries as the ID which is incorrect for this task. We will be changing those entries with the actual ID's by referring to the attr_type dataframe that shows which ID should be associated with a particular type of attraction.

To achieve this, we will define a function which will look for certain words that describe what type of attraction that is and then it will assign the particular ID for it.

```python
# Helper Function
def classify(name, id):
    name = name.lower()

    # For Temple category in AttractionTypeId column
    if any (x in name for x in ['cathedral', 'temple', 'mosque', 'shrine']):
        return 76    # Religious Sites
    

    # For Park category in AttractionTypeId column, there are few rows with national park, we will deal with them later
    # Here, we will deal with rows containing botanical garden, eco park and wildlife reserve, as these three come under the same ID
    if any (y in name for y in ['botanical garden', 'eco park', 'wildlife reserve']):
        return 63    # Nature & Wildlife Areas
    

    # For Museum category in AttractionTypeId column, there are rows that include
    # cultural heritage places, art gallery, history museum and science museum, we will deal with them accordingly
    # 1. Art Gallery and Science Museum
    if any (z in name for z in ['art gallery', 'science museum']):
        return 84    # Speciality Museums
    
    # 2. History Museum
    if 'history museum' in name:
        return 45    # History Museums
    
    # 3. Cultural Heritage Center
    if 'cultural heritage center' in name:
        return 72    # Points of Interest & Landmarks
    

    # For Market category in AttractionTypeId column
    if any (a in name for a in ['market', 'bazaar']):
        return 34    # Flea & Street Markets


    # For Beach category in AttractionTypeId column
    if any (b in name for b in ['beach', 'bay', 'cove']):
        return 13    # Beaches


    # For any remaining row that has National Park in it
    if 'national park' in name:
        return 61    # National Parks
    

    return id
```

```python
# Applying the function (also enforcing the dtype to be int64)
attr_str['AttractionTypeId'] = attr_str.apply(lambda c: classify(c['Attraction'], c['AttractionTypeId']), axis=1).astype(int)
```

Now that all the dataframes have been cleaned, we will merge them all together.

```python
# Merging
df = txn.merge(user[['UserId', 'CityId']], on='UserId', how='left')
df = df.merge(city, on='CityId', how='left')
df = df.merge(country, on='CountryId', how='left')
df = df.merge(region, on='RegionId', how='left')
df = df.merge(continent, on='ContinentId', how='left')
df = df.merge(attr_clean[['AttractionId', 'AttractionCityId', 'AttractionTypeId', 'Attraction']], on='AttractionId', how='left')
df = df.merge(attr_type, on='AttractionTypeId', how='left')
df = df.merge(mode, on='VisitModeId', how='left')

# Checking few rows of the merged dataframe
df.sample(7)
```

Despite earlier cleaning of all the dataframes before merging and also verifying that all dataframes had zero null values and the intended dtypes, we see that rows of few columns have null values and therefore to accomodate them the dtype has also changed from int64 to float64. This happened because the txn dataframe was the 'left' dataframe when it was merged with the user dataframe (how='left' was used) and before merging we removed the four rows with null values from user dataframe, so when merging, the rows of txn dataframe that had same common values as the ones removed due to cleaning the user dataframe, remained as it was, while the rows in the user dataframe for those values registered null values. So, to accomodate these null values, pandas, by default, changed the dtype of those numeric columns to float64.

We will clean this new merged dataframe now and save it in CSV format.

```python
# Removing the null values
df = df.dropna()

# Checking the shape and null values
print(df.shape)
print(df.isna().sum())
```

```python
# Converting the numeric columns to int64 dtype
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
df[numeric_columns] = df[numeric_columns].astype(int)

# Mapping the name of months to their respective numbers in the VisitMonth column
month_mapping = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'}
df['Month'] = df['VisitMonth'].map(month_mapping)

# Checking the data types after conversion
df.dtypes
```
